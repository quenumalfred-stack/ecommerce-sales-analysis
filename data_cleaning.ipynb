```python
df.columns
```

```python
df = df.dropna(subset=['Amount'])
```

```python
df = df.dropna(subset=['Qty'])
```

```python
df['Revenue'] = df['Qty'] * df['Amount']
```

```python
total_revenue = df['Amount'].sum()
print(total_revenue)
```

```python
df.columns
```

```python
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.info()
```

```python
df = df.drop(columns=["Unnamed: 22"])
```

```python
df = df.dropna(subset=["Amount"])
```

```python
monthly_sales = df.groupby(df["Date"].dt.to_period("M"))["Amount"].sum()

monthly_sales.sort_index()
```

```python
monthly_sales.plot(figsize=(10,5))
```

```python
top_categories = df.groupby("Category")["Amount"].sum().sort_values(ascending=False)

top_categories.head(10)
```

```python
top_products = df.groupby("SKU")["Qty"].sum().sort_values(ascending=False)

top_products.head(10)
```

```python
country_sales = df.groupby("ship-country")["Amount"].sum().sort_values(ascending=False)

country_sales
```

```python
status_counts = df["Status"].value_counts()

status_counts
```

```python
avg_order_value = df["Amount"].mean()

print(avg_order_value)
```

```python
df.to_csv("clean_sales.csv", index=False)
```
